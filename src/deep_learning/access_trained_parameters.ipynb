```python
import tensorflow as tf
from tensorflow.keras import layers, models

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Define a simple neural network
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),                # Flatten 28x28 input to 1D vector
    layers.Dense(128, activation='relu'),                # Dense hidden layer with 128 units and ReLU
    layers.Dense(10, activation='softmax')               # Output layer with 10 units for classification
])

# Compile the model with optimizer, loss function, and metric
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model (dummy variables x_train, y_train should be defined)
model.fit(x_train, y_train, epochs=5)



# ValueError: not enough values to unpack (expected 2, got 0)

# Access weights and biases of the first Dense layer (index 1)
weights_1, biases_1 = model.layers[1].get_weights()

print("Layer 1 - Dense(128)")
print("Weights shape:", weights_1.shape)  # Expected shape: (784, 128)
print("Biases shape:", biases_1.shape)    # Expected shape: (128,)
print("Weights:\n", weights_1)
print("Biases:\n", biases_1)
```
